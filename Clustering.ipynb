# DBSCAN


```python
from sklearn.cluster import DBSCAN
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(100)
data = pd.DataFrame(x)
data.columns = ["f1","f2"]
data["class"]= y
data.head()
```

```python
plt.scatter(data["f1"],data["f2"],cmap = plt.cm.Accent)
```

```python
#using kmeans
model = KMeans(n_clusters = 2,max_iter = 600,algorithm = "auto")
model.fit(data.iloc[:,:2])
plt.scatter(data["f1"],data["f2"],c = model.labels_,cmap = plt.cm.RdYlGn_r)
```

```python
model.labels_
```

```python
#using hirerachical
cluster = AgglomerativeClustering(n_clusters = 2,affinity = 'euclidean',linkage = "complete")
plt.scatter(data["f1"],data["f2"],c = cluster.fit_predict(data.iloc[:,:2]),cmap = plt.cm.Accent)
```

```python
#DBSCAN
dbs = DBSCAN(min_samples = 2,eps = 0.2)
class_pr = dbs.fit_predict(data.iloc[:,:2])
plt.scatter(data["f1"],data["f2"],c = class_pr,cmap = plt.cm.Accent)
```

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
nc = [2,3,4,5,6,7,8,9]
iner= []
for i in nc:
    km = KMeans(i)
    km.fit(iris.iloc[:,:4])
    iner.append(km.inertia_)
```

```python
plt.plot(nc,iner,color = "black")
plt.scatter(nc,iner,color = "red")
plt.show()
```

```python
#silhouette Score
```

```python
from sklearn.metrics import silhouette_score as sil
```

```python
nc = [2,3,4,5,6,7,8,9]
sc_score = []
for i in nc:
    hc=AgglomerativeClustering(n_clusters = i,affinity = "euclidean",linkage = "complete")
    
    y = hc.fit_predict(iris.iloc[:,:4])
    sc_score.append(sil(iris.iloc[:,:4],y))
```

```python
plt.plot(nc,sc_score,color = "black")
plt.scatter(nc,sc_score,color = "red")
plt.show()
```
